```python
# 1. Instala a versão clássica e estável da ferramenta para evitar erros da nova atualização
!pip install great_expectations==0.15.50 pandas

import pandas as pd
import great_expectations as ge

# 2. Lê a sua base de dados usando o Pandas
print("Carregando os dados, aguarde um instante...\n")
df_pandas = pd.read_csv('olist_order_items_dataset.csv')

# Converte para o formato do Great Expectations para fazer os testes
df = ge.from_pandas(df_pandas)

print("--- INICIANDO AUDITORIA DE QUALIDADE DE DADOS ---\n")

# 3. Regra de Negócio 1: Garantir que todo registro tem um ID de pedido (Integridade)
check_order = df.expect_column_values_to_not_be_null('order_id')
print(f"[{'APROVADO' if check_order['success'] else 'REPROVADO'}] Teste de integridade: Pedidos sem ID (Nulls)")

# 4. Regra de Negócio 2: Garantir que não existem preços zerados ou negativos (Faturamento)
check_price = df.expect_column_values_to_be_between('price', min_value=0.01, max_value=50000.00)
print(f"[{'APROVADO' if check_price['success'] else 'REPROVADO'}] Teste financeiro: Preços válidos (> zero)")

print("\n--- RELATÓRIO FINAL GERADO COM SUCESSO ---")
```
